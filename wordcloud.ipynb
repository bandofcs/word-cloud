```python
# Import packages
import re
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
#Read from text files
with open('input20.txt', 'r') as file:
    text = file.read()
# Clean text
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', '')
stop_words = ["brought","still","another","every","even","come","already","take","making","see","look","who","whose","must","many","well","let","us","will","national", "parade", "day","days"] + list(STOPWORDS)
```

```python
#Create a Singapore mask
mask = np.array(Image.open('sg.jpg'))
```

```python
# Generate wordcloud
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', colormap='rainbow', max_words=200, collocations=False, stopwords = stop_words, mask=mask).generate(text)
# Plot
plot_cloud(wordcloud)
```

```python
#Save to file
wordcloud.to_file("2020.png")
```
